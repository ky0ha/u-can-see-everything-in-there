目标：求$f(x)=x^2+5$的最大值，$x \in [0, 31]$

适应度函数为$f(x)=x^2+5$，初始种群大小为6，

```python
from collections import namedtuple


individual = namedtuple('individual', ['bins', 'value', 'fitness'])
```

```python
# init_population = ['01101', '00100', '01000', '10011', '00010', '00101']
from random import randint

init_population = []
for _ in range(500):
    value = randint(0, 31)
    init_population.append(individual(f"{bin(value)[2:]:0>5s}", value, value**2+5))
population = init_population
```
