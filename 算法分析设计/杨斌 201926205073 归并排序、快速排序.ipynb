```python
def loacl_sort(l, a, n):
    basic = n
    i, j = a, n
    while i!=j:
        while l[i]<=l[basic] and i!=j:
            i+=1
        while l[j]>=l[basic] and i!=j:
            j-=1
        if i!=j:
            l[i], l[j] = l[j], l[i]
        else:
            break
    
    l[i], l[basic] = l[basic], l[i]
    basic = i
    return basic

def quick_sort(l, a, n):
    if a<n:
        if n-a<=100:
            left, right = a, n
            while left<right:
                min, max = left, right
                for i in range(left, right+1):
                    if l[i]<l[min]:
                        min = i
                    if l[i]>l[max]:
                        max = i
                
                if l[min]<l[left]:
                    l[min], l[left] = l[left], l[min]
                if max==left:
                    max = min
                if l[max]>l[right]:
                    l[max], l[right] = l[right], l[max]
                
                right-=1
                left+=1
        else:
            pi = loacl_sort(l, a, n)

            quick_sort(l, a, pi-1)
            quick_sort(l, pi+1, n)

import time
from random import randint

array = [randint(1, 10**9) for _ in range(10**3)]

start = time.perf_counter()
quick_sort(array, 0, len(array)-1)
end = time.perf_counter()

print(f"排序耗时：{end-start}")
```

```python
def MergeSort(lists):
    if len(lists) <= 1:
        return lists
    num = int( len(lists) / 2 )
    left = MergeSort(lists[:num])
    right = MergeSort(lists[num:])
    return Merge(left, right)

def Merge(left,right):
    r, l=0, 0
    result=[]
    while l<len(left) and r<len(right):
        if left[l] <= right[r]:
            result.append(left[l])
            l += 1
        else:
            result.append(right[r])
            r += 1
    result += list(left[l:])
    result += list(right[r:])
    return result

import time
from random import randint

array = [randint(1, 10**9) for _ in range(10**3)]

start = time.perf_counter()
MergeSort(array)
end = time.perf_counter()

print(f"排序耗时：{end-start}")
```
